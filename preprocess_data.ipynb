## Crawling_data 합치기

```python
name_list = ['jsakommall_t','jsanjinmed_t','jshanemall_t','jshanpuremall_t','jshumanherb_t','jskmmall_t','jsoneshop_t']
```

```python
path = 'data/crawling_data/'
d1 = pd.read_json(path + f'{name_list[0]}.json')
d2 = pd.read_json(path + f'{name_list[1]}.json')
d3 = pd.read_json(path + f'{name_list[2]}.json')
d4 = pd.read_json(path + f'{name_list[3]}.json')
d5 = pd.read_json(path + f'{name_list[4]}.json')
d6 = pd.read_json(path + f'{name_list[5]}.json')
d7 = pd.read_json(path + f'{name_list[6]}.json')
```

```python
df = pd.concat([d1,d2,d3,d4,d5,d6,d7])
```

```python
len(df.id)
```

```python
df = df.iloc[df.id.drop_duplicates().index]
```

```python
df.index = range(len(df))
```

```python
df
```

```python
df.to_json('data/crawling_data/all.json',orient='records')
```
